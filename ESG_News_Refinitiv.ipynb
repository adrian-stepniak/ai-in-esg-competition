```python
companies_list = ['PB.L', 'EQBR.OL', 'HPCL.NS', 'BPCL.NS', 'DCC.L', 'PBF.N', 'ZEL.NZ', 'MUSA.N', 'PLNG.NS', 'SXY.AX',
                  'MAHM.NS', '7267.T', 'BMWG.DE', 'TAMO.NS', '7203.T', '7201.T', 'EPED.PA', 'SHA_p.DE', '7211.T',
                  '7261.T']
```

```python
gov_con_df = pd.DataFrame(columns=['companyCode', 'storyText'])
cyb_df = pd.DataFrame(columns=['companyCode', 'storyText'])
ceo_df = pd.DataFrame(columns=['companyCode', 'storyText'])
env_df = pd.DataFrame(columns=['companyCode', 'storyText'])
```

```python
def getting_headlines(df, topics, name):
    for company in companies_list:
        try:
            time.sleep(0.25)
            for topic in topics:
                for date in dates:
                    df = df.append(
                        ek.get_news_headlines('{} IN Language:LEN IN R:{}'.format(topic, company), date_from=date[0],
                                              date_to=date[1], count=100))
                    time.sleep(0.25)
                time.sleep(0.25)
            df['companyCode'].fillna(company, inplace=True)
        except:
            pass
    df = df.drop_duplicates()
    df.describe()
    df.to_csv('{}.csv'.format(name), sep=';')
    return df
```

```python
def get_news(df, name):
    for index, headlines_row in df.iterrows():
        try:
            story = ek.get_news_story(headlines_row['storyId'])
            soup = BeautifulSoup(story, "lxml")
            story_text = soup.get_text()
            df.at[index, 'storyText'] = story_text.split('}')[1:]
            time.sleep(0.25)
        except:
            pass
    df.to_csv('{}.csv'.format(name), sep=';')
```
